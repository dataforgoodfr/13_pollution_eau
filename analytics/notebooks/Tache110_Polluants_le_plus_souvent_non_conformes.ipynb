###  I. Etudier le nombre de prélèvements en non conformité 

**Quels sont les prélévelment qui dépassent la limite de qualité ?**

Dans le document PDF "eau-distrib-documentation"
 
#### 1.2.14 CONFORMITE DU PRELEVEMENT
La conformité du prélèvement est définie par rapport à deux types d’exigences de qualité, les limites de qualité et les références de qualité (cf § « les exigences de qualité » dans la partie « Contexte et description des données »), en distinguant pour chaque type d’exigence les paramètres microbiologiques des substances chimiques.

##### 1.2.14.1 plvconformitebacterio
Indicateur de la conformité des paramètres microbiologiques aux limites de qualité en vigueur au moment du prélèvement pour le type d’eau considéré.
Valeurs possibles : "blanc", "C=conforme", "N=non conforme", "S (sans objet lorsqu'aucun paramètre microbio n'a été mesuré)".

##### 1.2.14.2 plvconformitechimique
Indicateur de la conformité des paramètres chimiques aux limites de qualité en vigueur au moment du prélèvement pour le type d’eau considéré (et en prenant en compte les dérogations éventuelles en cours pour l'installation concernée).
Valeurs possibles : "blanc", "C=conforme", "N=non conforme", "D=conforme dans le cadre d’une dérogation","S (sans objet lorsqu'aucun paramètre chimique n'a été mesuré)".

##### 1.2.14.2.1 plvconformitereferencebact
Indicateur de la conformité des paramètres microbiologiques aux références de qualité en vigueur au moment du prélèvement pour le type d’eau considéré.
Valeurs possibles : "blanc", "C=conforme", "N=non conforme", "S (sans objet lorsqu'aucun paramètre microbio n'a été mesuré)".

##### 1.2.14.3 plvconformitereferencechim
Indicateur de la conformité des paramètres chimiques aux références de qualité en vigueur au moment du prélèvement pour le type d’eau considéré.
Valeurs possibles : "blanc", "C=conforme", "N=non conforme", "S (sans objet lorsqu'aucun paramètre chimique n'a été mesuré)".


```python
# [TO RUN] Compter les prélévements non conformes en 2024
query_2024_count_conformity = f"""
SELECT
  COUNT(*) AS nb_prel_tot,
  SUM(CASE 
        WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
        THEN 1 ELSE 0 END ) AS nb_prel_non_conformes,
  ROUND(SUM(CASE 
        WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
        THEN 1 ELSE 0  END )/ COUNT(*) 
        * 100,
        2) AS percent_prel_non_conformes_vs_all,
FROM edc_prelevements
WHERE
  dateprel >= '{date_min}'
"""

con.sql(query_2024_count_conformity).show()
```

###  II. Etudier la répartition géographique de ces prélèvements non conformes.

#### 1. Quelles sont les communes avec le plus de dépassements? 

Dans le document PDF "eau-distrib-documentation"

- **1.2.2 INSEECOMMUNEPRINC**: Code INSEE de la commune où se trouve le point de surveillance sur lequel le prélèvement a eu lieu.
- **1.2.3 NOMCOMMUNEPRINC**: Nom de la commune où se trouve le point de surveillance sur lequel le prélèvement a eu lieu.

```python
#  [TO RUN] Répartir les prélévements non conformes en 2024 par communes

query_2024_count_conformity_by_com_2 = f"""
SELECT
  NOMCOMMUNEPRINC, 
  INSEECOMMUNEPRINC,
  COUNT(*) AS nb_prel_tot,
  SUM(CASE 
        WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
        THEN 1 ELSE 0 END ) AS nb_prel_non_conformes,
  ROUND(SUM(CASE 
        WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
        THEN 1 ELSE 0  END )/ COUNT(*) 
        * 100,
        2) AS percent_prel_non_conformes
FROM edc_prelevements
WHERE
  dateprel >= '{date_min}'
GROUP BY 
   NOMCOMMUNEPRINC, 
   INSEECOMMUNEPRINC
ORDER BY 
   nb_prel_non_conformes  DESC,
   percent_prel_non_conformes DESC
"""

con.sql(query_2024_count_conformity_by_com_2).show()
```

```python
#  [TO RUN] Répartir les prélévements non conformes en 2024 par communes

query_2024_count_conformity_by_com = f"""
SELECT
  NOMCOMMUNEPRINC, 
  INSEECOMMUNEPRINC,
  COUNT(*) AS nb_prel_tot,
  SUM(CASE 
        WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
        THEN 1 ELSE 0 END ) AS nb_prel_non_conformes,
  ROUND(SUM(CASE 
        WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
        THEN 1 ELSE 0  END )/ COUNT(*) 
        * 100,
        2) AS percent_prel_non_conformes
FROM edc_prelevements
WHERE
  dateprel >= '{date_min}'
GROUP BY 
   NOMCOMMUNEPRINC, 
   INSEECOMMUNEPRINC
ORDER BY 
   percent_prel_non_conformes DESC,
   nb_prel_non_conformes  DESC
"""

con.sql(query_2024_count_conformity_by_com).show()
```

```python
#  [TO RUN] Répartir les prélévements non conformes en 2024 par commune

query_2024_conformity_com_repartition = f"""
WITH COMMUNES AS (
    {query_2024_count_conformity_by_com}
    ),
    
    TOTAL AS (
    SELECT
        COUNT(INSEECOMMUNEPRINC) AS tot_commune
    FROM
        COMMUNES
    )
    
SELECT
    (CASE WHEN percent_prel_non_conformes = 100 THEN '100%'
         WHEN percent_prel_non_conformes <10 THEN '0-10%'
         WHEN percent_prel_non_conformes <20 THEN '11-20%'
         WHEN percent_prel_non_conformes <30 THEN '21-30%'
         WHEN percent_prel_non_conformes <40 THEN '31-40%'
         WHEN percent_prel_non_conformes <50 THEN '41-50%'
         WHEN percent_prel_non_conformes <60 THEN '51-60%'
         WHEN percent_prel_non_conformes <70 THEN '61-70%'
         WHEN percent_prel_non_conformes <80 THEN '71-80%'
         WHEN percent_prel_non_conformes <90 THEN '81-90%'
         WHEN percent_prel_non_conformes <100 THEN '91-99%'
         END) AS percent_prel_non_conformes_cat, 
    COUNT(INSEECOMMUNEPRINC) AS nb_communes,
    MAX(tot_commune) AS tot_commune,
    ROUND(COUNT(INSEECOMMUNEPRINC)/MAX(tot_commune) *100,2) AS percent_communes
FROM
    COMMUNES
CROSS JOIN 
    TOTAL
GROUP BY
    percent_prel_non_conformes_cat
ORDER BY 
    percent_prel_non_conformes_cat
"""

con.sql(query_2024_conformity_com_repartition).show()
```

####  2. Dans quels département sont elles situées? 

Cette donnée est disponible dans edc_prelevements

```python
# [TO RUN] Répartir les prélévements non conformes en 2024 par département

query_2024_count_conformity_by_dep = f"""
SELECT
  cddept AS cddept,
  COUNT(*) AS nb_prel_tot,
  SUM(CASE 
        WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
        THEN 1 ELSE 0 END ) AS nb_prel_non_conformes,
  ROUND(SUM(CASE 
        WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
        THEN 1 ELSE 0  END )/ COUNT(*) 
        * 100,
        2) AS percent_prel_non_conformes
FROM 
    edc_prelevements
WHERE
  dateprel >= '{date_min}'
GROUP BY 
   cddept
ORDER BY 
   percent_prel_non_conformes DESC,
   nb_prel_non_conformes  DESC
"""

con.sql(query_2024_count_conformity_by_dep).show()
```

```python
# [TO RUN] Répartir les prélévements non conformes en 2024 par département

query_2024_conformity_dep_repartition = f"""
WITH DEP AS (
    SELECT
      cddept,
      COUNT(*) AS nb_prel_tot,
      SUM(CASE 
            WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
            THEN 1 ELSE 0 END ) AS nb_prel_non_conformes,
      ROUND(SUM(CASE 
            WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
            THEN 1 ELSE 0  END )/ COUNT(*) 
            * 100,
            2) AS percent_prel_non_conformes
    FROM 
        edc_prelevements
    WHERE
      dateprel >= '{date_min}'
    GROUP BY 
       cddept
    ),
    
    TOTAL AS (
    SELECT
        COUNT(cddept) AS tot_dept
    FROM
        DEP
    )
    
SELECT
    (CASE WHEN percent_prel_non_conformes = 100 THEN '100%'
         WHEN percent_prel_non_conformes <10 THEN '0-10%'
         WHEN percent_prel_non_conformes <20 THEN '11-20%'
         WHEN percent_prel_non_conformes <30 THEN '21-30%'
         WHEN percent_prel_non_conformes <40 THEN '31-40%'
         WHEN percent_prel_non_conformes <50 THEN '41-50%'
         WHEN percent_prel_non_conformes <60 THEN '51-60%'
         WHEN percent_prel_non_conformes <70 THEN '61-70%'
         WHEN percent_prel_non_conformes <80 THEN '71-80%'
         WHEN percent_prel_non_conformes <90 THEN '81-90%'
         WHEN percent_prel_non_conformes <100 THEN '91-99%'
         END) AS percent_prel_non_conformes_cat, 
    COUNT(cddept) AS nb_dept,
    MAX(tot_dept) AS tot_dept,
    ROUND(COUNT(cddept)/MAX(tot_dept) *100,2) AS percent_dept
FROM
    DEP
CROSS JOIN 
    TOTAL
GROUP BY
    percent_prel_non_conformes_cat
ORDER BY 
    percent_prel_non_conformes_cat
"""

con.sql(query_2024_conformity_dep_repartition).show()
```

```python
# [TO RUN] Répartir les prélévements non conformes en 2024 par region

query_2024_conformity_reg_repartition = f"""
WITH 
REGION_INFO AS (
    {query_REGION_INFO}
),
REG AS (
    SELECT
      REGION_INFO.name_region AS name_region,
      COUNT(*) AS nb_prel_tot,
      SUM(CASE 
            WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
            THEN 1 ELSE 0 END ) AS nb_prel_non_conformes,
      ROUND(SUM(CASE 
            WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
            THEN 1 ELSE 0  END )/ COUNT(*) 
            * 100,
            2) AS percent_prel_non_conformes
    FROM 
        edc_prelevements
    LEFT JOIN 
        REGION_INFO
    ON
        REGION_INFO.DEP = edc_prelevements.cddept
    WHERE
      dateprel >= '{date_min}'
    GROUP BY 
       REGION_INFO.name_region
    ),
    
    TOTAL AS (
    SELECT
        COUNT(name_region) AS tot_region
    FROM
        REG
    )
    
SELECT
    (CASE WHEN percent_prel_non_conformes = 100 THEN '100%'
         WHEN percent_prel_non_conformes <10 THEN '0-10%'
         WHEN percent_prel_non_conformes <20 THEN '11-20%'
         WHEN percent_prel_non_conformes <30 THEN '21-30%'
         WHEN percent_prel_non_conformes <40 THEN '31-40%'
         WHEN percent_prel_non_conformes <50 THEN '41-50%'
         WHEN percent_prel_non_conformes <60 THEN '51-60%'
         WHEN percent_prel_non_conformes <70 THEN '61-70%'
         WHEN percent_prel_non_conformes <80 THEN '71-80%'
         WHEN percent_prel_non_conformes <90 THEN '81-90%'
         WHEN percent_prel_non_conformes <100 THEN '91-99%'
         END) AS percent_prel_non_conformes_cat, 
    COUNT(name_region) AS nb_regions,
    MAX(tot_region) AS tot_region,
    ROUND(COUNT(name_region)/MAX(tot_region) *100,2) AS percent_regions
FROM
    REG
CROSS JOIN 
    TOTAL
GROUP BY
    percent_prel_non_conformes_cat
ORDER BY 
    percent_prel_non_conformes_cat
"""

con.sql(query_2024_conformity_reg_repartition).show()
```

```python
# [OPTIONAL]

query_check_refqual = """ 
WITH CAT AS (
    SELECT
      libmajparametre,
      categorie,
    FROM
      mapping_categories
)

SELECT 
  CAT.categorie AS cat_polluant,
  CAT.libmajparametre AS libmajparametre,
  edc_resultats.limitequal,
  edc_resultats.refqual
FROM  
    edc_resultats 
LEFT JOIN
    CAT
ON 
  edc_resultats.libmajparametre =   CAT.libmajparametre
GROUP BY
    1,2,3,4
HAVING 
    cat_polluant IS NOT NULL
    AND refqual IS NOT NULL   
    AND cat_polluant != 'non classé'
"""

con.sql(query_check_refqual).show()
```

Il n'y a pas toujours de `refqual` ou de `limitequal` dans `edc_resultats`<br>

- _1.3.1.6 Limitequal_ : Limite(s) de qualité du paramètre concerné en vigueur au moment du prélèvement pour le type d’eau considéré.
- _1.3.1.7 Refqual_ : Référence(s ) de qualité du paramètre concerné en vigueur au moment du prélèvement pour le type d’eau considéré.

Reprenons le travail fait dans la tache 132 : analytics/notebooks/Tache132_Etude_limite_de_qualite.ipynb

```python
# [TO RUN]

query_limit = """ 
WITH CAT AS (
    SELECT
      *
    FROM
      mapping_categories
),
LIM AS (
    SELECT 
      cdparametre,
      libmajparametre,
      cdparametresiseeaux,
      casparam,
      STRING_AGG(DISTINCT limitequal) AS limitequal,
      STRING_AGG(DISTINCT CAST(regexp_extract(REPLACE("limitequal", ',', '.'), '-?\d+(\.\d+)?') AS FLOAT)) AS limitequal_float,
      STRING_AGG(DISTINCT regexp_extract("limitequal", '[a-zA-Zµg]+/?[a-zA-Z/L]+$')) AS limitequal_unite,
      COUNT(DISTINCT limitequal) As nb_limitequal
    FROM  
        edc_resultats 
    GROUP BY 
      cdparametre,
      libmajparametre,
      cdparametresiseeaux,
      casparam
)
SELECT 
  CAT.*,
  LIM.limitequal,
  LIM.limitequal_float,
  LIM.limitequal_unite,
  COALESCE(LIM.nb_limitequal,0) AS nb_limitequal
FROM  
    CAT 
LEFT JOIN
    LIM
ON 
  LIM.cdparametre =   CAT.cdparametre
  AND LIM.libmajparametre = CAT.libmajparametre
  AND  LIM.cdparametresiseeaux = CAT.cdparametresiseeaux
  AND  LIM.casparam = CAT.casparam
"""
```

```python
# [TO RUN]
query_polluant = f""" 
WITH 
CAT AS (
      {query_limit}
),

RESULATS AS (
    SELECT
      edc_resultats.referenceprel,
      edc_resultats.libmajparametre,
      CAT.categorie AS categorie,
      (CASE WHEN valtraduite > CAST(CAT.limitequal_float AS FLOAT) THEN 1 ELSE 0 END) AS is_non_conforme
    FROM  
        edc_resultats 
    LEFT JOIN
        CAT
    ON 
      edc_resultats.cdparametre =   CAT.cdparametre
      AND edc_resultats.libmajparametre = CAT.libmajparametre
      AND  edc_resultats.cdparametresiseeaux = CAT.cdparametresiseeaux
      AND  edc_resultats.casparam = CAT.casparam
    WHERE 
        nb_limitequal = 1
    )
    
    SELECT
      categorie ,
      COUNT(*) AS nb_test_tot,
      SUM(CASE WHEN is_non_conforme = 1 THEN 1 ELSE 0 END ) AS nb_prel_non_conformes,
      ROUND(SUM(CASE WHEN is_non_conforme = 1 THEN 1 ELSE 0 END  )/ COUNT(*)  * 100,2) AS percent_prel_non_conformes_vs_all,
    FROM 
        RESULATS
    GROUP BY 
        categorie 
    ORDER BY
        percent_prel_non_conformes_vs_all  DESC
"""

con.sql(query_polluant).show()
```
