```python
# recherche liste unique referenceprel

edc_plv = data_edc_plv[["referenceprel"]].copy()
unique_edc_plv = edc_plv.drop_duplicates()
unique_edc_plv.info()
```

```python
# recherche des prélèvements présents dans EDR et absents dans EDC

resultat = unique_edr_plv[
    ~unique_edr_plv["referenceprel"].isin(unique_edc_plv["referenceprel"])
]
print(resultat)
```

```python
# extraction des 8588 prélèvements absents de EDC

res = resultat.merge(
    edr_udi_plv, left_on="referenceprel", right_on="referenceprel", how="inner"
)
res.describe(include="all")
```
